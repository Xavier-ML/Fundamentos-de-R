2. Graficar supervivencia por género

```python
import matplotlib.pyplot as plt
datos_supervivencia_genero = datos[["Survived","Sex"]]
datos_supervivencia_genero.head(5)
datos_supervivencia_genero.dtypes
```

```python
datos_supervivencia_genero
```

```python
var1 = datos_supervivencia_genero.groupby("Sex")["Survived"].count()
var1
```

```python
var1.head(10).plot.barh()
plt.xlabel('Promedio supervivencia')
plt.ylabel('Género')
plt.title('Distribución de género por supervivencia')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(datos[datos['Survived'] == 0]['Sex'], bins=5, label='No Sobrevivió', alpha=0.6, color='red')
ax.hist(datos[datos['Survived'] == 1]['Sex'], bins=5, label='Sobrevivió', alpha=0.6, color='green')

ax.set_title('Sobrevivientes por edad')
ax.set_xlabel('Edad')
ax.set_ylabel('Cantidad de personas')

ax.legend(loc='upper right')

plt.show()
```

9. Box plot de rango de edades de sobrevivientes

```python
import seaborn as sns
bins = [0, 10, 20, 30, 40, 50, 60, 70]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70']
datos['AgeGroup'] = pd.cut(datos['Age'], bins=bins, labels=labels, right=False)

supervivientes = datos[datos['Survived'] == 1]


plt.figure(figsize=(10, 6))
sns.boxplot(x='AgeGroup', y='Age', data=supervivientes, palette='Set3')
plt.title('Gráfico box plot de rangos de edades de supervivientes')
plt.xlabel('Grupo de Edad')
plt.ylabel('Edad')
plt.grid(True, linestyle='--', alpha=0.7)  # Agregar una grilla
plt.show()
```
